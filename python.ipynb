```python
def fac_recursive(n):
    if n == 1:
        print("1: 1!")
        return 1
    else:
        print(f"{n}! = {n} * {n-1}!")
        return n * fac_recursive(n - 1)

fac_recursive(5)
```

```python
def fac_recursive(n):
    if n == 1:
        print("1: 1!")
        return 1
    else:
        print(f"{n}! = {n} * {n-1}!")
        n_m = fac_recursive(n - 1)
        print(f"{n}! = {n} * {n_m}")
        return n * n_m

fac_recursive(5)
```
